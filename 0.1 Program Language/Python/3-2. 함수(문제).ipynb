## [소수 구별](https://freedeveloper.tistory.com/391)
프로그래머 X는 입력값을 숫자를 입력하거나 문자를 입력하려고 하는데,
- 만약 숫자를 입력하였으면 그것이 정수인지, 소수인지 구별하는 프로그램을 짜보도록 하고,
- 만약 문자를 입력하였으면 숫자가 아니므로 math error를 표시하게 하라.

### 정답

```python
def check_decimal(p_answer):
    is_decimal = True
    result = ""
    for i in range(2, p_answer):
        # x가 해당 수로 나누어떨어진다면
        if p_answer % i == 0:
            result = "소수가 아닙니다."
            is_decimal = False
            break

    if is_decimal:
      result = "소수입니다."

    return result
```

```python
def run():
    result = ""
    try:
        answer = input("숫자를 입력해주세요: ")
        answer = int(answer)
        result = check_decimal(answer)
    except:
        result = "math error"
    finally:
        print(result)
```

```python
run()
```

## [최대값, 최소값, 중간값 출력](https://codingdojang.com/scode/708#answer-filter-area)
- 예시 >> [8, 6, 2, 8, 5, 1, 2, 5, 4, 9]
  - 최소값: 1 / 최대값: 9 / 중간값: 5



### 정답

```python
import random

def make_random_list(p_range):
    lst = None
    try:
        lst = [ random.randint(0,9) for _ in range(p_range)]
        random.shuffle(lst)
    except Exception as e:
        print(f'[Error] make_random_list: {str(e)}')

    return lst
```

```python
def run(p_range=10):
    lst = make_random_list(p_range)
    if not lst:
        return

    print(f'make_random_list: {lst}')
    lst.sort() # 정렬
    print(f'after lst: {lst}')

    median = (len(lst) -1) / 2
    minimum = lst[0]
    maximum = lst[-1]

    print(f'최소값: {minimum} / 최대값: {maximum} / 중간값: {lst[int(median)]}')
```

```python
run('5')
```

```python
run(5)
```

```python
run()
```

## [Up&Down 숫자 맞추기](https://codingdojang.com/scode/711#answer-filter-area)
- 컴퓨터가 1 ~ 100 중 랜덤으로 숫자를 정합니다.
- 입력한 숫자보다 정답이 크면, "UP" 출력,
- 입력한 숫자보다 정답이 작으면, "DOWN" 출력,
- 정답을 맞추면, "정답" 출력

### 정답

```python
def fnc_ckeck(answer, correct, trial_num):
    result = False
    if not answer.isdigit():
        print("숫자만 입력해주세요.")
        return result

    answer = int(answer)
    if  answer == correct:
        print(f"정답입니다. 입력한 횟수: {trial_num}")
        result = True
    elif answer > correct:
        print("DOWN")
    else:
        print("UP")

    return result
```

```python
import random

def run(p_range=100):
    try:
        trial_num = 0
        correct = random.randint(1,p_range)
        print(f'correct: {correct}')

        while True:
            trial_num += 1
            answer = input("1~100 중 하나의 숫자를 입력해주세요 : ")
            if fnc_ckeck(answer, correct, trial_num):
                break
    except Exception as e:
        print(f'Error: {str(e)}')
```

```python
run('50')
```

```python
run(50)
```

## [숫자 세기 알고리즘](https://codingdojang.com/scode/720#answer-filter-area)
- 0부터 9999까지 8을 포함하지 않은 수는 총 몇개일까?
  - 8, 108, 888, 998 등은 8을 포함하는 수입니다.
  - 111, 299, 437 등은 8을 포함하지 않은 수입니다.

### 정답

```python
def run(num, include=False):
    try:
        if include:
            lst = [ i for i in range(10000) if str(i).count(str(num)) ]
        else:
            lst = [ i for i in range(10000) if not str(i).count(str(num)) ]

        print(len(lst))
    except Exception as e:
        print(f'Error: {str(e)}')
```

- 출력:
  - 구분자(Separator)는 마침표 '.', 쉼표 ',', 공백 ' ' 이다.
  - 가장 많이 나온 순서대로 단어 10개와 그 단어의 빈도를 출력한다.
  - 빈도가 같은 단어들 사이의 순서는 무시한다.
> { in : 12, the: 10, Blikango: 5 ... }

### 정답

```python
def make_context():
    context = "As the country became embroiled in a domestic crisis, the first government was dislodged and succeeded by several different administrations. Bolikango served as Deputy Prime Minister in one of the new governments before a partial state of stability was reestablished in 1961. He mediated between warring factions in the Congo and briefly served once again as Deputy Prime Minister in 1962 before returning to the parliamentary opposition. After Joseph-Desire Mobutu took power in 1965, Bolikango became a minister in his government. Mobutu soon dismissed him but appointed him to the political bureau of the Mouvement Populaire de la Revolution. Bolikango left the bureau in 1970. He left Parliament in 1975 and died seven years later. His grandson created the Jean Bolikango Foundation in his memory to promote social progress. The President of the Congo posthumously awarded Bolikango a medal in 2005 for his long career in public service."
    return context
```

```python
def make_words(p_context):
    words = p_context.replace('.', ' ').replace(',', ' ').split(' ')
    words = [ word for word in words if word ]

    return words
```

```python
from collections import Counter

def run(p_context=None, most_k=10):
    result = None
    try:
        if not p_context:
            p_context = make_context()
        elif not isinstance(p_context, str):
            result = f'is not context >> {p_context}'
            return result

        words = make_words(p_context)

        most_common = Counter(words).most_common(most_k)
        result = dict(most_common)
    except Exception as e:
        result = str(e)

    return result
```

## [가위 바위 보](https://codingdojang.com/scode/643#answer-filter-area)
- 사용자 입력과 random함수를 사용하여, 사용자(User)와 컴퓨터가 대결하는 가위 바위 보 게임을 만들어보자.
- 입력: [문자열] "가위", "바위" 혹은 "보" 출력: [문자열] 결과 반환

### 정답

```python
def init():
    승리의경우 = [('가위', '보'), ('바위', '가위'), ('보', '바위')]
    가위바이보 = ['가위', '바위', '보']
    컴퓨터 = random.choice(가위바이보)

    return 가위바이보, 승리의경우, 컴퓨터
```

```python
def who_is_winner(유저, 컴퓨터, 승리의경우):
    if 유저 == 컴퓨터:
        print("무승부입니다.")
    elif (유저, 컴퓨터) in 승리의경우:
        print("승")
    else:
        print("패")
```

```python
def run():
    가위바이보, 승리의경우, 컴퓨터 = init()
    print(f'컴퓨터: {컴퓨터}')

    유저 = input("가위바위보 중 하나를 입력해주세요: ")
    유저 = 유저.replace(' ','')
    if 유저 not in 가위바이보:
        print("잘 못 입력하였습니다.")

    who_is_winner(유저, 컴퓨터, 승리의경우)
```

## [남은 퇴근시간 계산하기](https://codingdojang.com/scode/645#answer-filter-area)
- input: 현재시간
- 퇴근시간: 18:00
- output: 남은시간 : hh:mm:ss

### 정답

```python
import datetime as dt
from pytz import timezone

def run(p_hour=18, p_minute=0):
    try:
        cur_time = dt.datetime.now(timezone('Asia/Seoul'))  # 현재시간
        leave_time = cur_time.replace(hour=p_hour, minute=p_minute, second=0)  # 퇴근시간

        # 남은 시간 계산 (퇴근 시간이 지난 경우 익일 퇴근까지의 시간)
        diffsec = (leave_time - cur_time).total_seconds()
        diffsec = diffsec + 60 * 60 * 24 if diffsec < 0 else diffsec

        print(f"다음 퇴근시간까지는 {dt.timedelta(seconds=diffsec)}({diffsec:,.0f}초) 남았습니다.")
    except Exception as e:
        print(f'Error: {str(e)}')
```

```python
run('a')
```

```python
run(15)
```

## [텍스트 파일 분류](https://codingdojang.com/scode/632#answer-filter-area)
- 랜덤으로 1, 2, 3 중 하나를 내용으로 갖는 txt 파일100개를 한 디렉토리 안에 생성하는 코드를 작성하시오. ( 파일 제목은 4자리 정수를 랜덤으로 할당. ex - 1382.txt , 0201.txt , 9012.txt , ......... )

- 제목이 0000-3333 인 txt 파일은 low 폴더로, 3334-6666인 txt 파일은 mid 폴더로, 6667~9999 인 파일은 high 폴더로 이동시키는 코드를 작성하세요. ( 폴더 생성도 코드 안에서 구현 )

- low, mid, high 폴더 안에 제목이 1, 2, 3 인 폴더를 각각 만들고, txt 파일 안의 내용에 따라 txt파일을 폴더안으로 이동시켜 분류하세요.

- 현재 디렉토리 - low - 1, 2, 3 ->1번폴더안의 Txt파일은 전부다 내용에 1 이라고 적혀있음. - mid - 1, 2, 3 - high - 1, 2, 3

```python
import shutil
import os

from random import randrange, choice
from math import ceil
from pathlib import Path
```

### 정답

```python
def mk_dir(init_folder_nm='sample'):
    sample_path = Path(init_folder_nm)

    # 폴더 생성
    if not sample_path.exists():
        sample_path.mkdir()

    # sub 폴더 생성
    for subdir in ['high', 'mid', 'low']:
        (sample_path/subdir).mkdir(exist_ok=True)

        subdirpath = sample_path/subdir
        for subsubdir in '123':
            (subdirpath/subsubdir).mkdir(exist_ok=True)

    return sample_path
```

```python
def mk_files(sample_path):
    # 파일 생성
    for _ in range(100):
        filename = '{:04}'.format(randrange(10000))+'.txt'
        Path(sample_path/filename).write_text(choice('123'))
```

```python
def move_by_file_nm(sample_path):

    for filepath in sample_path.glob('*.txt'):
        fn = int(str(filepath.relative_to(sample_path).with_suffix('')))
        subdir = 'low' if fn <= 3333 else ('mid' if fn <= 6666 else 'high')
        subdirpath = str(sample_path/subdir)

        subsubdir = filepath.read_text().strip(' \n\r')
        if subsubdir in list('123'):
            shutil.move(str(filepath), str(subdirpath+"/"+subsubdir))
        else:
            print(filepath, ': invalid file content')
```

```python
def run():
    # init
    sample_path = mk_dir()
    mk_files(sample_path)

    # move file
    move_by_file_nm(sample_path)
```

```python
run()
```

## [숫자야구](https://codingdojang.com/scode/721#answer-filter-area)
- 0-9까지의 숫자를 한 번씩 사용하여 만든 세 자리 숫자를 맞추는 코드를 작성하시오.
- 숫자와 자리가 맞으면 `S`이고, 숫자만 맞으면 `B`입니다.

컴퓨터가 세 자리 숫자를 설정하고 사용자에게 숫자를 입력받아 S와 B의 개수를 알려주십시오. 정답을 맞히면 정답이라고 알려주고 사용자가 숫자를 룰에 어긋나게 입력 시 경고문을 출력하고 다시 숫자를 입력할 수 있게 하십시오.
- 예) 정답이 123일 때,
  - 사용자가 234를 입력 시 0S2B,
  - 사용자가 109를 입력 시 1S0B,
  - 사용자가 327을 입력 시 1S1B입니다.


### 정답

```python
import random

def mk_corrects():
    nums = list(range(0,10))
    random.shuffle(nums)

    return [ str(i) for i in nums[:3]]
```

```python
def check_answers(p_answers):
    error_msg = ""
    if not p_answers.isdigit():
        error_msg = "잘못된 입력입니다. 숫자만 입력해주세요."
    elif len(p_answers) != 3:
        error_msg = "잘못된 입력입니다. 세자리 숫자를 입력해주세요."
    elif len(set(p_answers)) != 3:
        error_msg = "잘못된 입력입니다. 중복되지 않은 세자리 숫자를 입력해주세요."

    return error_msg
```

```python
def run(corrects=None):
    if not corrects:
        corrects = mk_corrects()
        print(f"corrects: {corrects}")

    answers = input("세자리 숫자를 입력하세요 : ")

    # 입력값 검증
    error_msg = check_answers(answers)
    if error_msg:
        print(error_msg)
        run(corrects)

    # 정답 확인
    elif list(answers) == corrects:
        print("정답을 맞추셨습니다. 게임을 종료합니다.")

    else:
        # 정답 비교
        strike = 0
        ball = 0
        for i in range(3):
            if answers[i] == corrects[i]:
                strike += 1
            elif answers.count(corrects[i]):
                ball += 1

        print(f'{strike}S{ball}B'.format(strike=strike, ball=ball))
        run(corrects)
```

```python
run()
```

## [베스킨라빈스31](https://codingdojang.com/scode/700#answer-filter-area)
- 게임의 참여자들은 차례를 정해 1부터 31까지의 수를 순차적으로 부른다.
- 한번에 1~3개까지 수를 연달아 부를 수 있으며, 마지막 31을 부른 사람이 진다.

힌트1:4n-2라는 공식을 사용하면 됩니다.     
힌트2:이 게임은 31을 부르면 지는 게임이니 30을 부르면 이깁니다.

### 정답

```python
import enum

class ERROR_MSG(enum.Enum):

    ERROR1 = ("not {0}.isdigit()", ["str_player"], "숫자 입력을 하지 않았습니다.", [])
    ERROR2 = ("not len({0})", ["player"], "입력을 하지 않았습니다.", [])
    ERROR3 = ("len({0}) > 3", ["player"], "최대 3개의 숫자만 입력 가능합니다. 입력한 숫자 수: {0}", ["len(player)"])
    ERROR4 = ("int({0}[0]) != ({1} + 1)", ["player", "COM"], "올바르게 입력하지 않았습니다. {0}부터 시작하는 숫자를 입력해주세요.", ["COM + 1"])
    ERROR5 = ("int({0}[-1]) > 31", ["player"], "올바르게 입력하지 않았습니다. 31보다 큰 숫자인 {0}는 입력할 수 없습니다.", ["player[-1]"])
    ERROR6 = ("len(set({0})) != len({1})", ["player", "player"], "올바르게 입력하지 않았습니다. 중복된 숫자는 입력할 수 없습니다. 입력된 숫자들: {0}", ["str_player"])
    ERROR7 = ("int({0}[-1]) != (int({1}[0])+(len({2})-1))", ["player", "player", "player"], "올바르게 입력하지 않았습니다. 차례대로 입려되지 않았습니다. 입력된 숫자들: {0}", ["str_player"])

    @classmethod
    def run_eval(self, player, COM):

        str_player = "".join(player)
        result = True
        for name in ERROR_MSG.__members__:

            _conditions = [ i for i in ERROR_MSG[name].value[1] ]
            if eval(ERROR_MSG[name].value[0].format(*_conditions)):

                _msgs = [eval(i) for i in ERROR_MSG[name].value[3]]
                print(ERROR_MSG[name].value[2].format(*_msgs))
                result = False
                break

        return result
```

```python
def run(turn=1, COM=0, player='0', is_ok=True):

    if is_ok:
        for i in range(int(player[-1]) + 1, (( 4 * turn ) - 2) + 1 ):
            print(i)
            COM = i

    player = input('콤마로 구분지어 숫자 입력: ').replace(' ','').split(',')
    player.sort()

    is_ok = ERROR_MSG.run_eval(player, COM)
    if is_ok:
        turn += 1

    if is_ok and COM == 30 :
        print('플레이어 패배!')
        return
    elif is_ok and int(player[-1]) == 30 :
        print('컴퓨터 패배!')
        return
    else:
        run(turn, COM, player, is_ok)
```
