## [소수 구별](https://freedeveloper.tistory.com/391)
프로그래머 X는 입력값을 숫자를 입력하거나 문자를 입력하려고 하는데,
- 만약 숫자를 입력하였으면 그것이 정수인지, 소수인지 구별하는 프로그램을 짜보도록 하고,
- 만약 문자를 입력하였으면 숫자가 아니므로 math error를 표시하게 하라.

### 정답

```python
class Game:

    def __check_decimal(self, p_answer):
        is_decimal = True
        result = ""

        for i in range(2, p_answer):
            # x가 해당 수로 나누어떨어진다면
            if p_answer % i == 0:
                result = "소수가 아닙니다."
                is_decimal = False
                break

        if is_decimal:
          result = "소수입니다."

        return result

    def run(self):

        result = ""
        try:

            answer = input("숫자를 입력해주세요: ")
            answer = int(answer)
            result = self.__check_decimal(answer)
        except:

            result = "math error"
        finally:

            print(result)
```

```python
game = Game()

game.run()
```

## [최대값, 최소값, 중간값 출력](https://codingdojang.com/scode/708#answer-filter-area)
- 예시 >> [8, 6, 2, 8, 5, 1, 2, 5, 4, 9]
  - 최소값: 1 / 최대값: 9 / 중간값: 5



### 정답

```python
import random
```

```python
class RandomList:
    def __init__(self, p_range=10):
        self.lst = [ random.randint(0,9) for _ in range(p_range)]

    def __call__(self):
        random.shuffle(self.lst)
        return self.lst
```

```python
class Game:
    def __init__(self):
        self.random_list = RandomList()
        self.median = 0
        self.minimum = 0
        self.maximum = 0

    def __process(self, lst):
        print(f'before list: {lst}')
        lst.sort() # 정렬
        print(f'after lst: {lst}')

        self.median = (len(lst) -1) / 2
        self.minimum = lst[0]
        self.maximum = lst[-1]

    def run(self, p_range=None):
        if p_range:
            self.random_list = RandomList(p_range)

        lst = self.random_list()
        self.__process(lst)

        print(f'최소값: {self.minimum} / 최대값: {self.maximum} / 중간값: {lst[int(self.median)]}')
```

```python
game = Game()

game.run()
```

```python
game.run(5)
```

## [Up&Down 숫자 맞추기](https://codingdojang.com/scode/711#answer-filter-area)
- 컴퓨터가 1 ~ 100 중 랜덤으로 숫자를 정합니다.
- 입력한 숫자보다 정답이 크면, "UP" 출력,
- 입력한 숫자보다 정답이 작으면, "DOWN" 출력,
- 정답을 맞추면, "정답" 출력

### 정답

```python
import random

class ParentGame:
    def __init__(self, p_range):
        self.p_range = p_range
        self.correct = 0
        self._mk_correct()

    def _mk_correct(self):
        self.correct = random.randint(1,self.p_range)
```

```python
class Game(ParentGame):
    def __init__(self, p_range=100):
        super().__init__(p_range)
        self.trial_num = 0
        self.answer = ''

    def __init(self):
        self.trial_num = 0
        self.answer = ''
        self._mk_correct()

    def __check(self):
        result = False

        if not self.answer.isdigit():
            print("숫자만 입력해주세요.")
            return result

        self.answer = int(self.answer)
        if  self.answer == self.correct:
            print(f"정답입니다. 입력한 횟수: {self.trial_num}")
            result = True
        elif self.answer > self.correct:
            print("DOWN")
        else:
            print("UP")

        return result

    def run(self):
        try:
            self.__init()
            print(f'correct: {self.correct}')

            while True:
                self.trial_num += 1
                self.answer = input("1~100 중 하나의 숫자를 입력해주세요 : ")
                if self.__check():
                    break
        except Exception as e:
            print(f'Error: {str(e)}')
```

```python
game = Game()

game.run()
```

```python
game.run()
```

## [방탈출 게임](https://codingdojang.com/scode/726#answer-filter-area)
- 비밀번호는 3자리로 한다.
- 비밀번호를 입력하게 한다.
- 비밀번호가 틀리면 처음부터 다시 입력하게 한다.
- 비밀번호가 맞으면 성공 메시지를 출력한다.

### 정답

```python
class MakePassword:
    def __init__(self):
        self.lst = [ str(i) for i in range(10) ]

    def _mk_password(self):
        _password = random.sample(self.lst, k=3)
        return "".join(_password)
```

```python
class ShowMSG:
    def _show_msg(self, answer, password):
      # 입력값 검증
      if len(answer) != 3:
          print("비밀번호 길이가 틀립니다.")
      elif not answer.isdigit():
          print("숫자만 입력해주세요.")

      # 정답 비교
      elif answer == password:
          print('성공')
      else:
          print('틀림')
```

```python
class Game(MakePassword, ShowMSG):
    def __call__(self):
        password = self._mk_password()
        print(f'password: {password}')
        answer = input()

        self._show_msg(answer, password)
```

```python
game = Game()

game()
```
