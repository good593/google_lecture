# [Homebrew](https://brew.sh/index_ko)
Homebrew는 Apple(또는 Linux 시스템)에서 제공하지 않는 유용한 패키지 관리자를 설치합니다.

- mac m1인 경우에는 brew 명령어 앞에 `arch -arm64`를 추가해줘야 합니다.
  > arch -arm64 brew upgrade  

image.png

## brew 명령어

- brew update:   
brew 최신 업데이트
- brew upgrade:    
설치된 항목들을 모두 업그레이드    
  - brew upgrade --greedy:    
  formula만 아니라 cash까지 한 번에 upgrade 진행
- brew search [설치할 항목명]:    
설치할 항목의 정확한 이름을 찾을 때 사용  
- brew install [설치할 정확한 이름]:
항목(패키지) 설치
- brew list:     
설치된 항목(패키지)들을 조회
- brew uninstall [삭제할 항목명]:    
설치된 패키지를 삭제할 때 사용
- brew cleanup:    
필요없는 tmp 파일들을 삭제
- brew outdated --greedy:    
brew update/upgrade를 하였을 때, 되지 않은 항목들을 확인할 수 있다.
- brew deps --tree --installed [패키지 명]:    
패키지 명의 종속된 dependencies들을 볼 수 있다.
- brew docter:    
homebrew와 관련된 문제가 발생할 경우 이 명령어를 통해 진달을 할 수 있습니다.
