# [Github](https://git-scm.com/)
> Git은 버전 관리 시스템이다.

버전 관리 시스템의 장점
- 원하는 버전 상태로 자유롭게 움직일 수 있다.
- 작업물의 변경 이력을 관리할 수 있다.
- 버전에 따라 수정 내용을 비교할 수 있다.
- 누가 문제를 일으켰는지 추적이 가능하다.
- 파일을 잃어버리거나 잘못 고쳤을 경우에도 쉽게 복구가 가능하다.
- 예상하지 못한 문제가 발생 시 빠르게 대처 가능하다.
