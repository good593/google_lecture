## 에디터

- `Ctrl + b`:     
왼쪽 탐색기 창 끄기/켜기
- `Ctrl + '+' / '-'`:    
폰트 사이즈 조절 
- `Ctrl + \`:    
현재 파일을 두 개의 세로 분할 패널로 나눠줌 
  - `Ctrl + 1 / 2`:    
  분할 패널에서의 포커스 이동(1:왼쪽, 2:오른쪽)
