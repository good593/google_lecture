# Object Detection
> 다수의 사물이 존재하는 상황에서 각 사물의 위치와 클래스를 찾는 작업

- Classification:  
한 이미지가 있을 때 그 클래스가 무엇인지 식별   
- Object Detection:   
한 이미지가 있을 때 그 안에 클래스가 무엇인지 + 그 위치는 어떻게 되는지 식별

image.png

## Object Detection 방법

- 1-Stage Detector:  
  - 물체의 위치를 찾는 문제(localization)과 분류(classification) 문제를 한 번에 해결하는 방식
  - 대표적인 예: YOLO

image.png

- 2-Stage Detector:
  - 물체의 위치를 찾는 문제(localization)과 분류(classification) 문제를 순차적으로 해결하는 방식
  - 대표적인 예: RCNN

image.png

# Region Proposal

## Sliding Window     
이미지에서 다양한 형태의 window를 슬라이딩하며 물체가 존해하는지 확인한다.

이러한 과정으로 인해 너무 많은 영역에 대하여 확인해야 한다는 단점이 있습니다. 그리고 이는 feature map이 아니라 입력 이미지에 대해서 cpu 장치를 이용해 sliding window를 진행하게 되면 넓은 input space 상에서 모두 탐색하게 되어 느릴 수 있습니다.

![img](https://blog.kakaocdn.net/dn/bHiDLF/btq7SMUlOw9/OjrUnQHvHvi3cijLbSDziK/img.gif)

## Selective Search     
인접한 영역(region)끼리 유사성을 측정하여 큰 영역으로 차례대로 통합해서 나가는 방식이다. 

Selective Search는 다음과 같은 순서로 진행된다.

1. 초기 sub-segmentation을 수행한다.

image.png

2. 작은 영역을 반복적으로 큰 영역으로 통합한다.

image.png

3. 통합된 영역들을 바탕으로 후보 영역을 만들어낸다.

image.png

## Region Proposal Network

RPN이 동자하는 알고리즘은 다음과 같습니다.  
1. CNN을 통해 뽑아낸 피쳐 맵을 입력으로 받습니다. 이때, 피처맵의 크기를 H x W x C로 잡습니다. (가로, 세로, 채널)
2. 피처맵에 3x3 컨블루션을 256 혹은 512 채널만큼 수행합니다.

# Faster RCNN
- 기존 RCNN과 Fast RCNN은 selective search 과정이 모두 CPU 내에서 이뤄지기 때문에 속도가 느릴 수 밖에 없었습니다.   
- Faster RCNN은 GPU 상에서 연산을 하기 위해 Region Proposal Network(RPN)을 제안하였다. 이는 Feature map을 보고 어느 곳에 물체가 있을지를 예측하는 것이다.
