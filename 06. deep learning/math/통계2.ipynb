# [이산확률 분포](https://wikidocs.net/34006) 
- 이산표본공간에서 정의된 확률변수의 값이 유한 혹은 countably infinite(자연수, 정수)하다.  
- 확률질량함수:    
이산확률변수 X의 값 x1, x2, ..., x_n의 각 확률을 대응  


## 이항분포(binomial distribution)
이항분포는 확률변수가 배타적인 두 가지 범주를 갖고 각 시행은 독립적인 경우에서의 분포입니다. 이 시행을 `베르누이 시행`이라고도 합니다.

총 n번의 독립적인 베르누이시행에서 관심 있는 범주가 나올 확률이 p라고 한다면,     
이런 조건들이 이항분포를 결정짓는 '상황'이며,      
여기서 이항분포는 관심있는 범주가 나오는 횟수 y를 확률변수로하는 분포입니다.

$$
Y \sim Bin(n,p) 이면,
$$
$$
P(Y=y) = {n \choose y}p^y(1 - p)^{n - y} \qquad y=0, 1, 2, ... , n
$$
$$
{n \choose y} = { n! \over y!(n-y)! } = _{n}\mathrm{C}_{y}(이항계수)
$$

## 포아송분포(poisson distribution)
- 이항분포: n번 중 성공확률이 p일 때, 성공 횟수를 확률변수로 하는 분포 
- 포아송분포: 일정 단위(시간)에서 평균 성공 수가 m(기댓값)일 때, 성공 횟수를 확률변수로 하는 분포 

예를 들어 어떤 공장에서 10시간(일정 단위)마다 평균적으로 2개의 불량품(평균 성공 수 m)이 발생된다면 불량품이 하나도 발생하지 않을 확률부터 수십, 수백개가 발생할 확률까지 성공 횟수에 따른 확률을 다루는 분포입니다.

$$
Y \sim Poisson(m)이면, 
$$
$$
p(Y = y) = { e^{-m}m^y \over y! } \qquad y=0,1,2, ...
$$
- $e$: 자연상수
- $E(Y) = m$: Y에 대한 평균 값 
- $V(Y) = m$: Y에 대한 분산 값 

포아송분포는 특이하게도 평균과 분산이 같습니다. 그래서 빈도 데이터에 적용하기가 적절합니다.

## [정규분포(normal distribution)](https://moni-gram.tistory.com/m/5) 
정규분포는 주어진 데이터의 형태가 산봉우리일 때 가정되는 분포입니다. 흔히 '종모양'에 비유하기도 합니다. 

$$
X \sim N(\mu, \sigma^2)이면,
$$
$$
f(x) = { 1 \over \sigma \sqrt{2\pi} } \exp \left( { -(x - \mu)^2  \over 2\sigma^2 } \right)  \qquad   - \infty < x < \infty
$$
- $E(X) = \mu$: 평균      
데이터의 중심 
- $V(X) = \sigma^2$: 분산($\sigma$: 표준편차)     
데이터의 퍼짐 정도

### [표준정규분포](https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=padosori60&logNo=220735271219)
- 표준정규분포에 대한 정의는 모든 정규분포를 평균($\mu$)은 0, 표준편차($\sigma$)는 1이 되도록 표준화한 것이다.
- 어떤 관측값 X의 값이 그 분포의 평균으로부터 표준편차의 몇 배정도나 떨어져 있는 가를 표준화된 정규분포 확률변수 Z로 나타내기 때문에 `Z분포`라고도 한다.
$$
Z = {X - \mu \over \sigma }
$$

### [중심극한정리](https://bioinformaticsandme.tistory.com/277) 
- 평균이 $\mu$, 표준편차 $\sigma$인 임의의 모집단으로 부터 크기 n인 표본에서의 표본평균은 n이 크면 근사적으로 평균이 $\mu$이고 분산이 ${\sigma^2 \over n}$인 정규분포를 따른다. 
- 모집단이 정규분포라면 표본평균은 표본 n개수와 상관없이 항상 정규분포를 따른다.

## [카이제곱($x^2$) 분포](https://angeloyeo.github.io/2021/12/13/chi_square.html)
양의 정수 k에 대해 k개의 독립적이고 표준정규분포를 따르는 확률변수 $X_1, X_2, ...., X_k$를 정의하면 자유도 k의 카이제곱 분포는 확률변수의 분포이다.  
$$
Q = \sum{X_i}^2 = X_1^2 + X_2^2 + \cdots + X_k^2 
$$ 

image.png

- 자유도에 따른 카이제곱 분포의 형태    
자유도가 커질 수록 정규분포에 가까워진다. 

image.png

### 카이제곱 분포의 의미 
- 카이제곱 분포는 오차(error) 혹은 편차(deviation)를 분석할 때 도움을 받을 수 있는 분포이다. 
- 중심극한정리에 따라 샘플수가 무수히 많고 합(sum)을 이용해 오차를 정의하는 경우 오차의 분포는 정규분포를 따르게 된다. 
- 그러므로 이렇듯 오차 혹은 편차에 대한 분석을 수행할 때 카이제곱 분포를 이용하면 이 오차 혹은 편차가 우연히 발생할 수 있다고 볼 수 있을만한 수준의 것인지 그렇지 않은지를 판별할 수 있게 되는 것이다. 

### 카이제곱 검정 
> 적합도 검정(goodness-of-fit test)와 교차 분석(cross tabulation analysis)을 이용할 때 카이제곱 분포를 이용하는 경우가 많다.     

피어슨 카이제곱 통계량
$$
x^2 = \sum{ (O_i - E_i)^2 \over E_i }
$$
- $O_i$: i번째 데이터에 대한 관찰값 
- $E_i$: i번째 데이터에 대한 기댓값 
- $(O_i - E_i)^2$: 편차의 제곱 
- ${ (O_i - E_i)^2 \over E_i }$: 특정 정규분포를 따름
