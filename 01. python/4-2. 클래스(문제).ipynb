## [소수 구별](https://freedeveloper.tistory.com/391)
프로그래머 X는 입력값을 숫자를 입력하거나 문자를 입력하려고 하는데,
- 만약 숫자를 입력하였으면 그것이 정수인지, 소수인지 구별하는 프로그램을 짜보도록 하고,
- 만약 문자를 입력하였으면 숫자가 아니므로 math error를 표시하게 하라.

### 정답

````python
is_decimal = True

try:
    answer = input("숫자를 입력해주세요: ")

    answer = int(answer)
    for i in range(2, answer):
        # x가 해당 수로 나누어떨어진다면
        if answer % i == 0:
            print("소수가 아닙니다.")
            is_decimal = False
            break

    if is_decimal:
      print("소수입니다.")
except:
    print("math error")
````

## [Up&Down 숫자 맞추기](https://codingdojang.com/scode/711#answer-filter-area)
- 컴퓨터가 1 ~ 100 중 랜덤으로 숫자를 정합니다.
- 입력한 숫자보다 정답이 크면, "UP" 출력,
- 입력한 숫자보다 정답이 작으면, "DOWN" 출력,
- 정답을 맞추면, "정답" 출력

### 정답

````python
import random

correct = random.randint(1,100)
correct
````

````python
trial_num = 0 # 시도 횟수

while True:
    trial_num += 1
    answer = input("1~100 중 하나의 숫자를 입력해주세요 : ")

    # 입력값 검증
    if not answer.isdigit():
        print("숫자만 입력해주세요.")
        continue

    answer = int(answer)
    if  answer == correct:
        print(f"정답입니다. 입력한 횟수: {trial_num}")
        break
    elif answer > correct:
        print("DOWN")
        continue
    else:
        print("UP")
        continue
````

## [가위 바위 보](https://codingdojang.com/scode/643#answer-filter-area)
- 사용자 입력과 random함수를 사용하여, 사용자(User)와 컴퓨터가 대결하는 가위 바위 보 게임을 만들어보자.
- 입력: [문자열] "가위", "바위" 혹은 "보" 출력: [문자열] 결과 반환

### 정답

````python
import random

가위바위보게임 = {
    '가위': {
        '바위':'패',
        '보':'승'
    },
    '바위': {
        '가위':'승',
        '보':'패'
    },
    '보': {
        '바위':'승',
        '가위':'패'
    }
}

가위바이보 = ['가위', '바위', '보']
컴퓨터 = random.choice(가위바이보)

answer = input("가위바위보 중 하나를 입력해주세요: ")
answer = answer.replace(' ','')
if answer not in 가위바이보:
    print("잘 못 입력하였습니다.")
elif answer == 컴퓨터:
    print("무승부입니다.")
else:
    print(가위바위보게임[answer][컴퓨터])

print(f'컴퓨터: {컴퓨터}')
````

## [숫자 정렬](https://codingdojang.com/scode/717#answer-filter-area)
- 자연수 리스트(홀수와 짝수의 개수가 같음)를 정렬해야 합니다.
  - 순서는 홀,짝,홀,짝 ... 으로 정렬
  - 홀수는 오름차순 정렬
  - 짝수는 내림차순 정렬

예) 4 1 3 2 6 5 -> 1 6 3 4 5 2

### 정답
- https://m.blog.naver.com/passion053/221074254811
- https://hcnoh.github.io/2019-01-27-python-arguments-asterisk

````python
from itertools import chain

answer = '123456'

even = [] # 짝수
odd = [] # 홀수

for i in answer:
    if int(i)%2:
        odd.append(i)
    else:
        even.append(i)

if len(even) != len(odd):
    print('홀수와 짝수의 개수가 다릅니다.')
else:
    even_sorted = sorted(even, reverse=True)
    odd_sorted = sorted(odd)
    print(*chain(*zip(odd_sorted, even_sorted)))
````

## [남은 퇴근시간 계산하기](https://codingdojang.com/scode/645#answer-filter-area)
- input: 현재시간
- 퇴근시간: 18:00
- output: 남은시간 : hh:mm:ss

### 정답

- 타임존을 설정하지 않으면, 기본값이 utc로 계산된다.

````python
import datetime as dt

cur_time = dt.datetime.now()  # 현재시간
leave_time = cur_time.replace(hour=18, minute=00, second=0)  # 퇴근시간
# 남은 시간 계산 (퇴근 시간이 지난 경우 익일 퇴근까지의 시간)
diffsec = (leave_time - cur_time).total_seconds()
diffsec = diffsec + 60 * 60 * 24 if diffsec < 0 else diffsec

print(f"다음 퇴근시간까지는 {dt.timedelta(seconds=diffsec)}({diffsec:,.0f}초) 남았습니다.")
````

- 한국 타임존으로 설정을 하여 퇴근시간이 잘 나온 것을 확인할 수 있다.

````python
import datetime as dt
from pytz import timezone

cur_time = dt.datetime.now(timezone('Asia/Seoul'))  # 현재시간
leave_time = cur_time.replace(hour=18, minute=00, second=0)  # 퇴근시간
# 남은 시간 계산 (퇴근 시간이 지난 경우 익일 퇴근까지의 시간)
diffsec = (leave_time - cur_time).total_seconds()
diffsec = diffsec + 60 * 60 * 24 if diffsec < 0 else diffsec

print(f"다음 퇴근시간까지는 {dt.timedelta(seconds=diffsec)}({diffsec:,.0f}초) 남았습니다.")
````

## [텍스트 파일 분류](https://codingdojang.com/scode/632#answer-filter-area)
- 랜덤으로 1, 2, 3 중 하나를 내용으로 갖는 txt 파일100개를 한 디렉토리 안에 생성하는 코드를 작성하시오. ( 파일 제목은 4자리 정수를 랜덤으로 할당. ex - 1382.txt , 0201.txt , 9012.txt , ......... )

- 제목이 0000-3333 인 txt 파일은 low 폴더로, 3334-6666인 txt 파일은 mid 폴더로, 6667~9999 인 파일은 high 폴더로 이동시키는 코드를 작성하세요. ( 폴더 생성도 코드 안에서 구현 )

- low, mid, high 폴더 안에 제목이 1, 2, 3 인 폴더를 각각 만들고, txt 파일 안의 내용에 따라 txt파일을 폴더안으로 이동시켜 분류하세요.

- 현재 디렉토리 - low - 1, 2, 3 ->1번폴더안의 Txt파일은 전부다 내용에 1 이라고 적혀있음. - mid - 1, 2, 3 - high - 1, 2, 3

````python
from random import randrange, choice
from math import ceil
from pathlib import Path
import shutil
````

````python
sample_path = Path('sample')

# 폴더 생성
if not sample_path.exists():
    sample_path.mkdir()

# 파일 생성
for _ in range(100):
        filename = '{:04}'.format(randrange(10000))+'.txt'
        Path(sample_path/filename).write_text(choice('123'))

# sub 폴더 생성
for subdir in ['high', 'mid', 'low']:
    (sample_path/subdir).mkdir(exist_ok=True)
````

### 정답

````python
for filepath in sample_path.glob('*.txt'):
    fn = int(str(filepath.relative_to(sample_path).with_suffix('')))
    subdir = 'low' if fn <= 3333 else ('mid' if fn <= 6666 else 'high')
    shutil.move(str(filepath), str(sample_path/subdir)) # don't care overwriting
````

````python
for subdir in ['high', 'mid', 'low']:
    subdirpath = sample_path/subdir
    for subsubdir in '123':
        (subdirpath/subsubdir).mkdir(exist_ok=True)
````

````python
for filepath in subdirpath.glob('*.txt'):
        subsubdir = filepath.read_text().strip(' \n\r')
        if subsubdir in list('123'):
            shutil.move(str(filepath), str(subdirpath/subsubdir))
        else:
            print(filepath, ': invalid file content')
````

## [숫자야구](https://codingdojang.com/scode/721#answer-filter-area)
- 0-9까지의 숫자를 한 번씩 사용하여 만든 세 자리 숫자를 맞추는 코드를 작성하시오.
- 숫자와 자리가 맞으면 `S`이고, 숫자만 맞으면 `B`입니다.

컴퓨터가 세 자리 숫자를 설정하고 사용자에게 숫자를 입력받아 S와 B의 개수를 알려주십시오. 정답을 맞히면 정답이라고 알려주고 사용자가 숫자를 룰에 어긋나게 입력 시 경고문을 출력하고 다시 숫자를 입력할 수 있게 하십시오.
- 예) 정답이 123일 때,
  - 사용자가 234를 입력 시 0S2B,
  - 사용자가 109를 입력 시 1S0B,
  - 사용자가 327을 입력 시 1S1B입니다.


### 정답

````python
import random

nums = list(range(0,10))
random.shuffle(nums)

corrects = [ str(i) for i in nums[:3]]
corrects
````

````python
while True:
    answers = input("세자리 숫자를 입력하세요 : ")

    # 입력값 검증
    if not answers.isdigit():
        print("잘못된 입력입니다. 숫자만 입력해주세요.")
        continue
    elif len(answers) != 3:
        print("잘못된 입력입니다. 세자리 숫자를 입력해주세요.")
        continue
    elif len(set(answers)) != 3:
        print("잘못된 입력입니다. 중복되지 않은 세자리 숫자를 입력해주세요.")
        continue

    # 정답 확인
    elif list(answers) == corrects:
        print("정답을 맞추셨습니다. 게임을 종료합니다.")
        break

    # 정답 비교
    strike = 0
    ball = 0
    for i in range(3):
        if answers[i] == corrects[i]:
            strike += 1
        elif answers.count(corrects[i]):
            ball += 1

    print(f'{strike}S{ball}B'.format(strike=strike, ball=ball))
````

## [베스킨라빈스31](https://codingdojang.com/scode/700#answer-filter-area)
- 게임의 참여자들은 차례를 정해 1부터 31까지의 수를 순차적으로 부른다.
- 한번에 1~3개까지 수를 연달아 부를 수 있으며, 마지막 31을 부른 사람이 진다.

힌트1:4n-2라는 공식을 사용하면 됩니다.     
힌트2:이 게임은 31을 부르면 지는 게임이니 30을 부르면 이깁니다.

### 정답

````python

turn = 1
COM = ''
player = '0'
is_ok = True

while True:
    if is_ok:
        for i in range(int(player[-1]) + 1, (( 4 * turn ) - 2) + 1 ):
            print(i)
            COM = i
    player = input('콤마로 구분지어 숫자 입력: ').replace(' ','').split(',')
    player.sort()

    str_player = "".join(player)
    if not str_player.isdigit():
        print("숫자 입력을 하지 않았습니다.")
        is_ok = False
        continue
    elif not len(player):
        print("입력을 하지 않았습니다.")
        is_ok = False
        continue
    elif len(player) > 3:
        print(f"최대 3개의 숫자만 입력 가능합니다. 입력한 숫자 수: {len(player)}")
        is_ok = False
        continue
    elif int(player[0]) != (COM + 1):
        print(f"올바르게 입력하지 않았습니다. {COM + 1}부터 시작하는 숫자를 입력해주세요.")
        is_ok = False
        continue
    elif int(player[-1]) > 31:
        print(f"올바르게 입력하지 않았습니다. 31보다 큰 숫자인 {player[-1]}는 입력할 수 없습니다.")
        is_ok = False
        continue
    elif len(set(player)) != len(player):
        print(f"올바르게 입력하지 않았습니다. 중복된 숫자는 입력할 수 없습니다. 입력된 숫자들: {str_player}")
        is_ok = False
        continue
    elif int(player[-1]) != (int(player[0])+(len(player)-1)):
        print(f"올바르게 입력하지 않았습니다. 차례대로 입려되지 않았습니다. 입력된 숫자들: {str_player}")
        is_ok = False
        continue
    else:
        is_ok = True

    turn += 1
    if COM == 30 :
        print('플레이어 패배!')
        break
    elif int(player[-1]) == 30 :
        print('컴퓨터 패배!')
        break
````
