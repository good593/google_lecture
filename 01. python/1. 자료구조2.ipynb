## 집합 생성

- 순서가 없다.

````python
a = set([1, 1, 2, 2, 2])
a 
````

````python
type(a)
````

````python
b = set("Hello")
b
````

## 교집합

````python
a = set([1, 2, 3, 4, 5]) 
b = set([3, 4, 5, 6, 7])

a & b 
````

````python
a.intersection(b)
````

## 합집합

````python
a = set([1, 2, 3, 4, 5]) 
b = set([3, 4, 5, 6, 7])

a | b 
````

````python
a.union(b)
````

## 차집합

````python
a = set([1, 2, 3, 4, 5]) 
b = set([3, 4, 5, 6, 7])

a - b 
````

````python
a.difference(b)
````

## 추가

````python
a = set([1, 2, 3])
a.add('55')
a.add(77)
a.update(['a', 'b', 'c']) # 여러개 추가 
a
````

## 제거

# 불리언
불리언 또는 불(bool)이라고 하며, 참(True)과 거짓(False)을 나타내는 자료형이다.  

````python
a = True 
b = False 

a, b 
````

````python
type(a)
````

````python
not a # True and False -> False
````

````python
not b # False and False -> True 
````

````python
int(a) # True to 1
````

````python
int(b) # False to 0
````

## 자료형별 데이터의 참/거짓

### 숫자형

````python
a = 0 
b = 1
c = -1 
d = 3.14 

bool(a), bool(b), bool(c), bool(d)
````

### 문자형

````python
a = ""
b = "Hello world"

bool(a), bool(b)
````

### 리스트

````python
a = [] 
b = list() 
c = [1, 2, 3]
d = ['a', 'b']

bool(a), bool(b), bool(c), bool(d)
````

### 튜플

````python
a = ()
b = tuple()
c = (1,)
d = (1, 2, 3)

bool(a), bool(b), bool(c), bool(d)
````

### 딕셔너리

````python
a = {} 
b = dict() 
c = {'a':1}

bool(a), bool(b), bool(c)
````

### 집합

````python
a = set() 
b = set([1, 2, 3])
c = set(['a', 'b'])

bool(a), bool(b), bool(c)
````
